```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/salaries.csv')
```

```python
df.head()
```

```python
bins=[0, 20000, 40000, 60000, 80000, 100000, 120000, 140000, 160000, 180000, 200000, 220000, 240000, 260000, 280000, 300000, 320000, 340000, 360000, 380000, 400000]



plt.figure(figsize=(5, 3),dpi=300)



plt.hist(df.salary, bins=bins, color='blue', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
```

```python
top_jobs = ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer', 'Data Product Manager'] # Define top job titles to filter
job_counts = df['job_title'].value_counts().loc[top_jobs] # Filter for top jobs 



plt.bar(job_counts.index, job_counts.values, color='green', edgecolor='black') #to show the bar chart for top jobs
plt.title('Job Title Distribution')
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
# plt.show()
```

```python
top_jobs = ['AI Engineer', 'ML Scientist', 'Data Visualization Lead', 'Data Analytics Specialist'] # Define top job titles to filter
# job_counts = df['job_title'].value_counts().loc[top_jobs] # Filter for top jobs 
job_counts = df[df['job_title'].isin(top_jobs)]['salary_in_usd'].groupby(df['job_title']).mean().loc[top_jobs] # Calculate average salary for top jobs



plt.bar(job_counts.index, job_counts.values, color='green', edgecolor='black') #to show the bar chart for top jobs
plt.title('Job Title vs Average Salary')
plt.xlabel('Job Title')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=90)
# plt.show()
```
