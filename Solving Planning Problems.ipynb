```python
ParcelID = {
    "P1": (8, 9, 7, 6),
    "P2": (5, 6, 4, 3),
    "P3": (10, 10, 9, 8),
    "P4": (3, 4, 2, 5),
    "P5": (7, 8, 6, 7)
}

categories = {
    "Highly Suitable": [],
    "Moderately Suitable": [],
    "Not Suitable": []
}

for pid, scores in ParcelID.items():
    total = sum(scores)
    if total >= 30:
        categories["Highly Suitable"].append(pid)
    elif 20 <= total < 30:
        categories["Moderately Suitable"].append(pid)
    else:
        categories["Not Suitable"].append(pid)

for category, parcels in categories.items():
    print(f"{category}: {', '.join(parcels) if parcels else 'None'}")


```

```python
def get_top_applicants(applicants):
    # Calculate priority for each applicant
    scored_applicants = []
    for name, income, dependents, disability_status in applicants:
        disability_bonus = 500 if disability_status.lower() == "yes" else 0
        priority = (100000 - income) + (dependents * 2000) + disability_bonus
        scored_applicants.append((name, priority))
    # Sort by priority descending and get top 5 names
    top_applicants = sorted(scored_applicants, key=lambda x: x[1], reverse=True)[:5]
    return [name for name, _ in top_applicants]

# Example usage:
applicants = [
    ("Alice", 25000, 2, "no"),
    ("Bob", 40000, 3, "yes"),
    ("Charlie", 15000, 1, "no"),
    ("David", 50000, 4, "no"),
    ("Eve", 30000, 2, "yes"),
    ("Frank", 20000, 0, "no"),
    ("Grace", 35000, 5, "no")
]

top_5 = get_top_applicants(applicants)
print(top_5)
```

```python
# Function to check affordability of each zone
def check_affordability():
    zones = []

    # Take input for number of zones
    n = int(input("Enter number of city zones: "))

    # Input zone data
    for i in range(n):
        print(f"\nEnter details for Zone {i + 1}:")
        name = input("Zone name: ")
        income = float(input("Average monthly income (BDT): "))
        rent = float(input("Average monthly housing rent (BDT): "))
        ratio = rent / income

        affordable = "Yes" if ratio <= 0.3 else "No"

        zone_data = {
            "name": name,
            "income": income,
            "rent": rent,
            "ratio": ratio,
            "affordable": affordable
        }

        zones.append(zone_data)

    # Display report for each zone
    print("\n--- Zone Affordability Report ---")
    for zone in zones:
        print(f"\nZone: {zone['name']}")
        print(f"Income: {zone['income']} BDT")
        print(f"Rent: {zone['rent']} BDT")
        print(f"Rent-to-Income Ratio: {zone['ratio']:.2f}")
        print(f"Affordable: {zone['affordable']}")

    # Separate affordable and unaffordable zones
    affordable_zones = [z['name'] for z in zones if z['affordable'] == "Yes"]
    unaffordable_zones = [z['name'] for z in zones if z['affordable'] == "No"]

    # Display summary
    print("\nAffordable Zones:")
    print(", ".join(affordable_zones) if affordable_zones else "None")

    print("\nUnaffordable Zones:")
    print(", ".join(unaffordable_zones) if unaffordable_zones else "None")

# Run the program
check_affordability()
```

```python
import matplotlib.pyplot as plt

# Input number of zones
num_zones = int(input("Enter number of municipal zones: "))

zone_names = []
total_wastes = []

# Collect data and calculate total daily waste for each zone
for _ in range(num_zones):
    zone_name = input("Enter zone name: ")
    population = int(input("Enter population: "))
    waste_per_person = float(input("Enter daily waste generation rate per person (kg): "))
    total_waste = population * waste_per_person
    zone_names.append(zone_name)
    total_wastes.append(total_waste)

# Determine the zone with the highest total waste
max_waste = max(total_wastes)
max_zone = zone_names[total_wastes.index(max_waste)]
print(f"\nZone with highest total daily waste: {max_zone} ({max_waste:.2f} kg)")

# Plot bar chart
plt.figure(figsize=(8, 5))
plt.bar(zone_names, total_wastes, color='skyblue')
plt.xlabel('Zone')
plt.ylabel('Total Daily Waste (kg)')
plt.title('Total Daily Waste by Zone')
plt.tight_layout()
plt.show()
```

```python
import math

# City grid data
land_use = [['R', 'C', 'E', 'R'],
            ['R', 'R', 'C', 'E'],
            ['I', 'R', 'R', 'C'],
            ['E', 'C', 'I', 'R']]

population = [[120, 40, 0, 200],
              [100, 180, 30, 0],
              [80, 160, 140, 60],
              [0, 30, 50, 190]]

temperature = [[32, 36, 30, 34],
               [35, 37, 33, 31],
               [40, 38, 39, 35],
               [28, 30, 41, 36]]

# Bus stops coordinates
bus_stops = [(0, 0), (2, 3)]

# Service center coordinates
service_centers = [(0, 1), (2, 2)]

# Waste center coordinates
waste_center = (1, 0)

# Nearest Neighbor approach for TSP (simplified version)
def euclidean_distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

# Task 1: Zoning and Population Analysis
def show_zoning_summary():
    land_use_count = {'R': 0, 'C': 0, 'I': 0, 'E': 0}
    population_sum = {'R': 0, 'C': 0, 'I': 0, 'E': 0}
    total_population = 0

    for i in range(len(land_use)):
        for j in range(len(land_use[i])):
            land_type = land_use[i][j]
            pop = population[i][j]

            land_use_count[land_type] += 1
            population_sum[land_type] += pop
            total_population += pop

    for land_type in land_use_count:
        count = land_use_count[land_type]
        total_pop = population_sum[land_type]
        avg_pop = total_pop / count if count != 0 else 0
        population_percent = (total_pop / total_population) * 100 if total_population != 0 else 0

        print(f"{land_type}: {count} cells, Total Population: {total_pop}, Avg: {avg_pop:.2f}")
        print(f"Population in {land_type} zones: {total_pop} ({population_percent:.2f}% of total)")

# Task 2: Accessibility Analysis via Bus Stops
def check_bus_stop_accessibility():
    residential_cells = []
    for i in range(len(land_use)):
        for j in range(len(land_use[i])):
            if land_use[i][j] == 'R':  # Residential cell
                residential_cells.append((i, j))
                min_distance = float('inf')

                # Calculate distance to all bus stops
                for bus in bus_stops:
                    distance = euclidean_distance((i, j), bus)
                    min_distance = min(min_distance, distance)

                if min_distance > 1.5:
                    print(f"Residential cell at ({i},{j}) is more than 1.5 units away from a bus stop.")

    # Count the total population affected
    population_affected = 0
    for i, j in residential_cells:
        if land_use[i][j] == 'R' and min_distance > 1.5:
            population_affected += population[i][j]

    # Calculate percentage of affected population
    affected_percentage = (population_affected / total_population) * 100
    print(f"Population affected: {population_affected} ({affected_percentage:.2f}%)")

# Task 3: Waste Collection Route Coverage
def simulate_waste_collection():
    residential_cells = [(i, j) for i in range(len(land_use)) for j in range(len(land_use[i])) if land_use[i][j] == 'R']
    route = [waste_center]
    remaining_cells = residential_cells[:]
    total_distance = 0

    while remaining_cells:
        current = route[-1]
        nearest_cell = min(remaining_cells, key=lambda x: euclidean_distance(current, x))
        route.append(nearest_cell)
        remaining_cells.remove(nearest_cell)
        total_distance += euclidean_distance(current, nearest_cell)

    # Add distance back to waste center
    total_distance += euclidean_distance(route[-1], waste_center)

    print(f"Optimal Route: {route}")
    print(f"Total Distance: {total_distance:.2f} units")

# Task 4: Service Area Coverage (Hospitals, Fire Stations)
def analyze_service_area():
    far_cells = []
    for i in range(len(land_use)):
        for j in range(len(land_use[i])):
            if land_use[i][j] == 'R':  # Residential cell
                nearest_center = min(service_centers, key=lambda sc: euclidean_distance((i, j), sc))
                distance = euclidean_distance((i, j), nearest_center)
                if distance > 1.5:
                    far_cells.append((i, j))

    print(f"Cells more than 1.5 units away from service centers: {far_cells}")

# Task 5: Urban Heat Island Detection
def detect_uhi_risks():
    u_hi_risk_cells = []
    for i in range(len(land_use)):
        for j in range(len(land_use[i])):
            if (land_use[i][j] in ['R', 'C'] and temperature[i][j] > 36 and population[i][j] > 100):
                u_hi_risk_cells.append((i, j))

    print(f"UHI-risk cells: {u_hi_risk_cells}")

# Main menu for the terminal interface
def main():
    while True:
        print("Welcome to SmartCity Planner!")
        print("1. Show Zoning Summary")
        print("2. Check Bus Stop Accessibility")
        print("3. Simulate Waste Collection Routes")
        print("4. Analyze Service Area Coverage")
        print("5. Detect UHI Risks")
        print("6. Exit")
        
        choice = input("Select an option: ")
        
        if choice == '1':
            show_zoning_summary()
        elif choice == '2':
            check_bus_stop_accessibility()
        elif choice == '3':
            simulate_waste_collection()
        elif choice == '4':
            analyze_service_area()
        elif choice == '5':
            detect_uhi_risks()
        elif choice == '6':
            break
        else:
            print("Invalid choice. Please try again.")

# Run the program
if __name__ == "__main__":
    main()
```
