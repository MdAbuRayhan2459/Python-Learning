```python
#Check if Two Dictionaries are Equal
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 2, 'a': 1}
def are_dicts_equal(d1, d2):
    if len(d1) != len(d2):
        return False
    for key in d1:
        if key not in d2 or d1[key] != d2[key]:
            return False
    return True
print(are_dicts_equal(dict1, dict2))  # Output: True
```

```python
#Nested Dictionary Traversal
student = {
    "name": "Alice",
    "marks": {
        "math": 90,
        "science": 85
    }
}
for key, value in student.items():
    if isinstance(value, dict):
        print(f"{key}:")
        for sub_key, sub_value in value.items():
            print(f"  {sub_key}: {sub_value}")
    else:
        print(f"{key}: {value}")
        

```

```python
student = {
    "name": "Alice",
    "marks": {
        "math": 90,
        "science": {
            "theory": 80,
            "practical": 90
        }
    },
    "profile": {
        "age": 20,
        "contact": {
            "email": "alice@example.com",
            "phone": "123456"
        }
    }
}
def traverse_dict(d, indent=0):
    for key, value in d.items():
        print(' ' * indent + str(key) + ':', end=' ')
        if isinstance(value, dict):
            print()  # New line for nested dictionary
            traverse_dict(value, indent + 2)  # Increase indentation for nested dict
        else:
            print(value)
traverse_dict(student)
#Flatten a Nested Dictionary
```

```python
#Reverse Words in a String
def reverse_words(sentence):
    words = sentence.split()
    reversed_words = ' '.join(reversed(words))
    return reversed_words
sentence ="World Hello"
print(reverse_words(sentence))
```

```python
#Check Palindrome
def is_palindrome(s):
    s = s.lower().replace(" ", "")  # Normalize the string
    return s == s[::-1]  # Check if the string is equal to its reverse
s = "Was it a car or a cat I saw"
print(is_palindrome(s))  # Output: True
```

```python
print(is_palindrome("racecar"))  # Output: True
```

```python
#Find First Non-Repeating Character
def first_non_repeating_char(s):
    char_count = {}
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    for char in s:
        if char_count[char] == 1:
            return char
    return None
s = "swiss"
print(first_non_repeating_char(s))  # Output: 'w'
```

```python
#Capitalize Each Word
def capitalize_words(sentence):
    words = sentence.split()
    capitalized_words = [word.capitalize() for word in words]
    return ' '.join(capitalized_words)
sentence = "hello world"
print(capitalize_words(sentence))  # Output: "Hello World"
```

```python
#Count Occurrences of Each Character
def count_characters(s):
    char_count = {}
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count
s = "banana"
print(count_characters(s))  # Output: {'b': 1, 'a': 3, 'n': 2}
```

```python
#Check if Two Strings are Rotations
def are_rotations(s1, s2):
    if len(s1) != len(s2):
        return False
    return s1 in (s2 + s2) #s2 + s2 = "dabcdabc" whare s1 inside of (s2+s2)
s1 = "abcd"
s2 = "dabc"
print(are_rotations(s1, s2))  # Output: True
```

```python
#Remove Duplicate Characters
def remove_duplicates(s):
    seen = set()
    result = []
    for char in s:
        if char not in seen:
            seen.add(char)
            result.append(char)
    return ''.join(result)
s = "programming"
print(remove_duplicates(s))  # Output: "progamin"
```
