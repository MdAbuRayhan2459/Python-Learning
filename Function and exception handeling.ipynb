```python
def prime_numbers_checker(n):
    if n <= 1:
        return False
    for i in range(2, int(n+1)):
        if n % i == 0:
            return False
    return True
print(f"Is 5 a prime number? {prime_numbers_checker(5)}")
```

```python
#Scope of variables
x = 10  # Global variable
def show(): # Function to demonstrate variable scope
    x = 5  # Local variable
    print(x) 
show()
print(x)
```

```python
#Scope of variables
x = 10  # Global variable
def show(): # Function to demonstrate variable scope
    global x  # Declare x as global to modify it
    x = 5  # Local variable
    print(x) 
show()
print(x)
```

Exception Handling

```python
list1 = [1, 2,0, 3, 4, 5]

for i in list1:
    print(100/i)
```

```python
list1 = [1, 2,0, 3, 4, 5]

for i in list1:
    try:
        print(100/i)
    except ZeroDivisionError:
        print("Division by zero is not allowed.")
```

```python
#MULTIPLE EXCEPTIONS handling
list1 = [1, 2, 0, 3, 4, 'a', 5]
for i in list1:
    try:
        print(100 / i)
    except ZeroDivisionError:
        print("Division by zero is not allowed.")
    except TypeError:
        print("Invalid type for division. Please provide a number.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
#catching multiple exceptions in a single except block
list1 = [1, 2, 0, 3, 4, 'a', 5]
for i in list1:
    try:
        print(100 / i)
    except (ZeroDivisionError, TypeError) as e:
        print(f"An error occurred: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
def prime_numbers_checker(n):
    if n <= 1:
        return False
    for i in range(2, int(n+1)):
        if n % i == 0:
            return False
    return True
```

```python
def print_prime_numbers(num):
    if num<2:
        print("There are no prime numbers less than 2.")
        return
    for number in range(2,num+1):
        if prime_numbers_checker(number):
            print(number, end=' ')
    print()  # Print a newline after printing all prime numbers
print_prime_numbers(100) 
```
