```python
#factorial function
# factorial_numbers = int(input("Enter a number to calculate its factorial: "))
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1) 

    # print("Factorial of", factorial_numbers, "is", factorial(factorial_numbers))
```

```python
def is_armstrong_number(n):
    # Convert the number to a string to get its digits
    digits = [int(d) for d in str(n)]
    # Calculate the number of digits
    num_digits = len(digits)
    # Calculate the sum of digits raised to the power of num_digits
    armstrong_sum = sum(d ** num_digits for d in digits)
    # Check if the number is equal to the armstrong sum
    return n == armstrong_sum

# Example usage
n = 370
print(is_armstrong_number(n))  # Output: True
```

```python
for i in range(1, 1000):
    if is_armstrong_number(i):
        print(i)
```

```python
def print_diamond(n):
    # Upper part of the diamond
    for i in range(1, n + 1, 2):
        print(" " * ((n - i) // 2) + "*" * i)
    # Lower part of the diamond
    for i in range(n - 2, 0, -2):
        print(" " * ((n - i) // 2) + "*" * i)

# Example usage
print_diamond(5)
```
