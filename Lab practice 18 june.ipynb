```python
def palindromic_primes(start, end):
    def is_prime(n):
        if n <= 1:
            return False
        for i in range(2, int(n)):
            if n % i == 0:
                return False
        return True

    def is_palindrome(n):
        return str(n) == str(n)[::-1]

    palindromic_primes_list = []
    for num in range(start, end + 1):
        if is_prime(num) and is_palindrome(num):
            palindromic_primes_list.append(num)

    return palindromic_primes_list

```

```python
class Student:
    def __init__(self, name, id, marks=None):
        self.name = name
        self.id = id
        self.marks = marks if marks is not None else []

    def add_mark(self, mark):
        self.marks.append(mark)

    def average_mark(self):
        if not self.marks:
            return 0
        return sum(self.marks) / len(self.marks)
    def print_info(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        print(f"Marks: {self.marks}")
        print(f"Average Mark: {self.average_mark():.2f}")
```

```python
def is_prime(n):
    """Check if a number is prime."""
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def twin_primes(a, b):
    """Print all twin prime pairs between a and b (inclusive)."""
    for num in range(a, b-1):  # b-1 because we are checking (num, num+2)
        if is_prime(num) and is_prime(num + 2):
            if num + 2 <= b:
                print(f"({num}, {num + 2})")

# Example usage:
twin_primes(2, 50)
```

```python
def calculator(a, b, op):
    try:
        if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
            raise TypeError("Invalid data type")
        
        if op == '+':
            return a + b
        elif op == '-':
            return a - b
        elif op == '*':
            return a * b
        elif op == '/':
            if b == 0:
                return "Cannot divide by zero"
            return a / b
        else:
            return "Invalid operator"
    
    except TypeError:
        return "Invalid data type"
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
print(calculator(10, 5, '+'))  # Example usage
print(calculator(10, 5, '-'))  # Example usage
print(calculator(10, 5, '*'))  # Example usage
print(calculator(10, 5, '/'))  # Example usage
print(calculator(10, 0, '/') ) # Example usage with division by zeroprint(
print(calculator(10, '5', '+'))  # Example usage with invalid data type
```

```python
class BankAccount:
    def __init__(self, name, account_number):
        self.name = name
        self.account_number = account_number
        self.balance = 0.0  # Initial balance is zero
        print(f"Account created for {self.name} (Acc: {self.account_number})")

    def deposit(self, amount):
        if amount <= 0:
            print("Deposit amount must be positive.")
        else:
            self.balance += amount
            print(f"Deposited: {amount}. New Balance: {self.balance}")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive.")
        elif amount > self.balance:
            print("Insufficient balance.")
        else:
            self.balance -= amount
            print(f"Withdrew: {amount}. New Balance: {self.balance}")

    def get_balance(self):
        return self.balance

    def display_info(self):
        print("---- Account Info ----")
        print(f"Name: {self.name}")
        print(f"Account Number: {self.account_number}")
        print(f"Current Balance: {self.balance}")
        print("----------------------")
# Create first account
acc1 = BankAccount("Asifi", 1001)
acc1.deposit(500)
acc1.withdraw(200)
acc1.display_info()
print("Current Balance:", acc1.get_balance())
print()

# Create second account
acc2 = BankAccount("Rayhan", 1002)
acc2.deposit(1000)
acc2.withdraw(1200)  # Should show insufficient balance
acc2.withdraw(-100)  # Invalid withdrawal
acc2.deposit(-50)    # Invalid deposit
acc2.display_info()
print("Current Balance:", acc2.get_balance())
print()
```

```python
def safe_divide(a, b):
    """Safely divide two numbers with error handling."""
    try:
        if b == 0:
            raise ZeroDivisionError("Cannot divide by zero")
        return a / b
    except ZeroDivisionError as e:
        return str(e)
    except Exception as e:
        return f"An error occurred: {str(e)}"
# Example usage
print(safe_divide(10, 0))  # Should return 5.0
```

```python

class rectangle:
    def __init__(self,length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2 * (self.length + self.width)
    def display(self):
        print(f"Length: {self.length}, Width: {self.width}")
        print(f"Area: {self.area()}")
        print(f"Perimeter: {self.perimeter()}")
```

```python
rectangle1 = rectangle(5, 3)
rectangle1.display()
```

```python
class circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * (self.radius ** 2)

    def circumference(self):
        return 2 * 3.14 * self.radius

    def display(self):
        print(f"Radius: {self.radius}")
        print(f"Area: {self.area()}")
        print(f"Circumference: {self.circumference()}")
circle1 = circle(4)
circle1.display()
```

```python
class BankSystem:
    def __init__(self):
        self.accounts = []
    def create_account(self,name, account_number):
        for acc in self.accounts:
            if acc.account_number == account_number:
                print("Account already exists.")
                return
        new_acc = BankAccount(name, account_number)
        self.accounts.append(new_acc)
        print("account created successfully.")
    def find_account(self, account_number):
        for acc in self.accounts:
            if acc.account_number == account_number:
                return acc
        print("Account not found.")
        return None
    def transfer(self, from_acc_num, to_acc_num, amount):
        from_acc = self.find_account(from_acc_num)
        to_acc = self.find_account(to_acc_num)

        if from_acc is None or to_acc is None:
            
            print("Transfer failed: One or both accounts not found.")
            return

        if amount <= 0:
            print("Transfer amount must be positive.")
            return

        if from_acc.balance < amount:
            print("Transfer failed: Insufficient balance.")
            return

        from_acc.withdraw(amount)
        to_acc.deposit(amount)
        print("Transfer successful.")
    def display_all_accounts(self):
        print("=== All Accounts ===")
        for acc in self.accounts:
            acc.display_info()

```

```python
bank = BankSystem()

bank.create_account("Asifi", 1001)
bank.create_account("Rayhan", 1002)
bank.create_account("Tamim", 1003)

a1 = bank.find_account(1001)
a1.deposit(2000)

a2 = bank.find_account(1002)
a2.deposit(1000)

a3 = bank.find_account(1003)
a3.deposit(500)

bank.transfer(1002, 1003, 300)
bank.transfer(1001, 1002, 1500)  # Should show insufficient balance
bank.transfer(1003,1001,1000)
bank.display_all_accounts()
```
