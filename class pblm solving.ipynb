```python
def is_armstrong_number(n):
    # Convert the number to a string to get its digits
    digits = [int(d) for d in str(n)]
    # Calculate the number of digits
    num_digits = len(digits)
    # Calculate the sum of digits raised to the power of num_digits
    armstrong_sum = sum(d ** num_digits for d in digits)
    # Check if the number is equal to the armstrong sum
    return n == armstrong_sum

# Example usage
n = 370
print(is_armstrong_number(n))  # Output: True
```

```python
#hexadecimal to decimal conversion
hexadecimal = input("Enter a hexadecimal number:")
decimal = 0
power = 0
for digit in reversed(hexadecimal):
    if '0' <= digit <= '9':
        value = int(digit)
    elif 'A'<= digit <='F':
        value = ord(digit)-ord('A') +10
    else:
        print("Invalid hexadecimal digit:", digit)
        exit()
    decimal = decimal + value*(16 ** power)
    power = power + 1
print("Decimal value is:", decimal)
```

```python
def pascal_triangle(n):
    for i in range(n):
        # Print leading spaces for formatting
        print(" " * (n - i), end="")

        num = 1  # First value in a row is always 1
        for j in range(i + 1):
            print(num, end=" ")
            num = num * (i - j) // (j + 1)  # Calculate next value in the row
        print()  # Move to next line

# Input
rows = int(input("Enter number of rows: "))
pascal_triangle(rows)
```

```python
#Exercise 11: Print all prime numbers within a range
n = int(input("Enter a number:"))
for num in range(2,n+1):
    is_prime = True
    for i in range(2,num):
        if num%i == 0:
            is_prime = False
            break
    if is_prime:
        print(num,end=" ")
```

```python
#Exercise 17: Find the sum of a series of a number up to n terms
terms = 5
# first number of sequence
num = 2
sum_seq = 0

# run loop n times
for i in range(0, terms):
    print(num, end="+")
    sum_seq += num
    # calculate the next term
    num = num * 10 + 2
print("\nSum of above series is:", sum_seq)
```
